```python
def graficos_eda_categoricos(cat):
  #calculamos el numero de filas q necesitamos
    from math import ceil
    filas = ceil(cat.shape[1]/2)
  #definimos el grafico
    f,ax= plt.subplots(nrows= filas, ncols=2, figsize=(16,filas*6))
  #aplanamos para iterar por el grafico como si fuera 1 dimension en vez de 2
    ax= ax.flat
  #creamos el bucle q va creando los graficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax=ax[cada])
        ax[cada].set_title(variable,fontsize=12, fontweight= 'bold')
        ax[cada].tick_params(labelsize=12)
```

EDA VARIABLES NUMÉRICAS

```python
def estadisticos_cont(num):
    #calculamos describe
    estadisticos= num.describe().T
    #añadimos la mediana
    estadisticos['median']= num.median()
    #reordenamos para q la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    return (estadisticos)
```

```python
estadisticos_cont(df.select_dtypes('number'))
```

```python
#¿Cuál es el impacto economico de la fuga?segun un estudio americano:
#si el salario anual es <30K, un 16,1% de su sueldo
#si el salario anual es 30K-50k, un 19,7%
#si el salario anual es 50K-75k, un 20.4%
#si el salario anual es >75K, un 21%
#por lo tanto necesitamos calcular el salario anual de los empleados
```

```python
df3['salario_anual']=df3.salario_mes*12
df3['salario_anual']
```

```python
#calculamos el impacto economico de cada empleado
condiciones= [(df3['salario_anual'] <30000),
              (df3['salario_anual'] >30000) & (df3['salario_anual'] <50000),
              (df3['salario_anual']>50000)& (df3['salario_anual']<75000),
              (df3['salario_anual']>75000)]
resultados=[df3.salario_anual*0.161,df3.salario_anual*0.197,df3.salario_anual*0.204,df3.salario_anual*0.21]
df3['coste_abandono']=np.select(condiciones,resultados,default=-999)
df3
```

```python
#¿cuanto nos ha costado en el ultimo año el abandono?
coste_total=df3.loc[df3.abandono==1].coste_abandono.sum()
coste_total
```

```python
#¿cuanto nos cuesta q no esten motivados los empleados (motivacion baja)
df3.loc[(df3.abandono==1) & (df3.implicacion=='Baja')].coste_abandono.sum()
```

Primero vamos a calcular el % de representantes de ventas que se han ido el año pasado

```python
total_repre_pasado = len(df3.loc[df3.puesto == 'Sales Representative'])
abandonos_repre_pasado = len(df.loc[(df3.puesto == 'Sales Representative') & (df3.abandono == 1)])
porc_pasado = abandonos_repre_pasado / total_repre_pasado

porc_pasado
```

#ahora vamos a calcular cuantos se nos irán este año

```python
total_repre_actual = len(df3.loc[(df3.puesto == 'Sales Representative') & (df3.abandono == 0)])
se_iran = int(total_repre_actual * porc_pasado)

se_iran
```

```python
#Sobre ellos cuantos podemos retener (hipótesis 30%) y cuanto dinero puede suponer
retenemos = int(se_iran * 0.3)

ahorramos = df3.loc[(df3.puesto == 'Sales Representative') & (df3.abandono == 0),'coste_abandono'].sum() * porc_pasado * 0.3

print(f'Podemos retener {retenemos} representantes de ventas y ello supondría ahorrar {ahorramos}€.')
```

MODELO DE MACHINE LEARNING

```python
df_ml = df3.copy()
```

```python
df_ml.info()
```

```python
### PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN
```

# Transformar todas las variables categóricas a númericas


```python

from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_ml.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse_output = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Dataframe final

```python
#Seleccionamos las variables numéricas para poder juntarlas a las cat_hoe
num = df3.select_dtypes('number').reset_index(drop = True)
num
```

```python

df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

```python
### DISEÑO DE LA MODELIZACIÓN
#Separación predictoras y target
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

```python
#separación train y test
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

```python
#entrenamiento
from sklearn.tree import DecisionTreeClassifier

#Instanciar
ac = DecisionTreeClassifier(max_depth=4)

#Entrenar
ac.fit(train_x,train_y)
```

```python
# Predicción
pred = ac.predict_proba(test_x)[:, 1]
pred[:20]
```

```python
# Evaluación
from sklearn.metrics import roc_auc_score

roc_auc_score(test_y,pred)
```

INTERPRETACIÓN:DIAGRAMA DE ARBOL

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (50,50))

plot_tree(ac,
          feature_names= test_x.columns,
          impurity = False,
          node_ids = True,
          proportion = True,
          rounded = True,
          precision = 2);
```

```python
pd.Series(ac.feature_importances_,index = test_x.columns).sort_values(ascending = False).plot(kind = 'bar', figsize = (30,20));
```

```python
#Incoporación del scoring al dataframe principal
df3['scoring_abandono'] = ac.predict_proba(df_ml.drop(columns = 'abandono'))[:, 1]
df3
```

```python
#los 10 empleados con mayor probabilidad de dejar la empresa
df3.sort_values(by = 'scoring_abandono', ascending = False)[0:10]
```

```python
#riesgo de dejar la empresa por puesto de trabajo
df3.boxplot(column='scoring_abandono', by='puesto', figsize = (20,12));
```
